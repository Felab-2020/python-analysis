# Analysing the data 

Concatenating dataframes 

```
Telcom=pd.concat([database1,database2,database3])
print (Telcom)
```

Merging 

```
Telcomm= Telcom.merge(cg, left_on="SITE_ID", right_on ="SITE_CODE")
Telcomm
```

City with highest calls and sms

```
Telcomm[['PRODUCT', 'DATE_TIME', 'VALUE', 'VILLES', 'STATUS']]
Telcomm.sort_values(by=['PRODUCT'], inplace=True, ascending=False)
Telcomm
```
